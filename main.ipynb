<div id="imports">

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file and save the data as a Pandas DataFrame
space_df = pd.read_csv("Space_Corrected.csv")

# Make sure it was imported correctly by reading the head
space_df.head()
```

<div id="clean">
    <h3>Data Cleaning</h3>
</div>

```python
# Remove all NaN columns
space_df = space_df.dropna()

# Remove unamed columns since they are not helpful
space_df.drop(["Unnamed: 0.1", "Unnamed: 0"], axis = True, inplace = True)

# Read the head again of the DataFrame to confirm it worked
space_df.head()
```

<div id="processing">
    <h3>Data Processing</h3>
</div>

```python
# Make a year column out of the datum column
# We know years come right after the comma, so we split them and grab the first element
def grab_year(le_date):
    other, year_containing = le_date.split(",")
    return year_containing[1:5]

# Test that it works the way we want
year = grab_year(space_df["Datum"][0])
year
```

```python
# Generate a new pandas column called year out of this function
space_df["Year"] = space_df["Datum"].apply(grab_year)

# Check it worked with space_df.head(10) and space_df.tail(10)
space_df.head(10)
```

```python
space_df.tail(10)
```

```python
# Make another function to grab the country out of the location
# Split them by the comma and grab the last element which is the country
def grab_country(le_location):
    return le_location.split(",")[-1]

# Test it worked the way we want
country = grab_country(space_df["Location"][0])
country
```

```python
# Generate a new Pandas column called Country out of the function we created to extract the country
space_df["Country"] = space_df["Location"].apply(grab_country)
```

```python
# Test it worked with space_df.head(10) and space_df.tail(10)
space_df.head(10)
```

```python
space_df.tail(10)
```

```python
# Remove the location and Datum columns since we do not need them anymore, they contain too much detail
space_df.drop(["Location", "Datum"], axis = True, inplace = True)
```

```python
# Test it worked with space_df.head()
space_df.head()
```

<div id="visualization">
    <h3>Data Visualization</h3>
</div>

```python
# Make a bar chart that will show the percentage of success missions per country
def grab_success_percentage(specific_data_frame):

    # Grab the total number of missions of that country
    num_country_missions = len(specific_data_frame)
    
    # Grab the total number of successful missions of that country
    num_country_successful_missions = len(specific_data_frame[specific_data_frame["Status Mission"] == "Success"])

    try:
        # Calculate percentage
        return round((num_country_successful_missions / num_country_missions) * 100, 2)
    except ZeroDivisionError:
        return round(0, 2)
```

```python
# Grab percentages of all countries
countries = space_df["Country"].unique()

countries_success_percentages = []

for country in countries:

    # Grab country success percentage
    country_success_percentage = grab_success_percentage(space_df[space_df["Country"] == country])
    # Append it to the percentages list
    countries_success_percentages.append(country_success_percentage)

# Test out it worked by printing the percentages
countries_success_percentages
```

```python
# Generate a bar chart where the values are the success percentages percentages
colors = ['#5E4FA2', '#75BBB4', '#BA6C65', '#6F88A1', '#A1A8D4', '#C5D9EA', '#EBCC77', '#E7A070', '#6A0572', '#AB83A1', '#E3E3E3']
plt.figure(figsize=(20,15))
plt.title("Success Percentage of Missions per Country")
plt.xlabel("Country")
plt.ylabel("Success Mission Percentage")
plt.bar(countries, countries_success_percentages, color = colors, label = countries)
plt.legend(prop={"size": 15}, title="Countries")
plt.grid(axis = "y")
plt.show()
```

```python
# Now we do know the percentage of success, however we do not know the number of missions per country, let's find that out
num_missions_per_country = []

for country in countries:
    num_mission_per_country = len(space_df[space_df["Country"] == country])
    num_missions_per_country.append(num_mission_per_country)

num_missions_per_country
```

```python
# Now make a bar chart about the number of missions each country had
plt.figure(figsize=(20,15))
plt.title("Number of Missions per Country")
plt.xlabel("Country")
plt.ylabel("Number of Missions")
plt.bar(countries, num_missions_per_country, color = colors, label = countries)
plt.legend(prop={"size": 15}, title="Countries")
plt.grid(axis = "y")
plt.show()
```
